```python
import pandas as pd
import seaborn as sns
```

```python
avocado = pd.read_csv('/Users/roxannepostell/Downloads/avocado.csv')
avocado.head()
```

```python
avocado.info()
```

```python
avocado['Date'] = avocado['Date'].apply(pd.to_datetime)
avocado.info()
```

```python
tomato = avocado.query("Date >= '2015-01-01' and Date < '2016-01-01' and region == 'Albany' and type == 'conventional'").copy()
tomato.loc[:, ['Date', 'Total Bags', 'Small Bags']].head(5)
```

```python
potato = avocado.melt(
    id_vars = 'Date',
    value_vars=['Total Bags','Small Bags'],
    var_name ='Bags',
    value_name='Count')

with pd.option_context('display.max_rows', 10):
    display(potato)
```

```python
sns.lineplot(data=tomato, 
             x='Date', y='Count', hue='Bags')
```

```python
total_volume = avocado["Total Volume"]
pd.qcut(total_volume, q=4,
       labels=['Poor','Modest','Good','Excellent'])
avocado['Sales Volume'] = pd.qcut(total_volume, q=4,
       labels=['Poor','Modest','Good','Excellent'])
avocado.head(5)
```

```python
avocado['Year'] = pd.to_datetime(avocado['Date']).dt.strftime('%Y')
```

```python
sns.countplot(data=avocado, x='Year', hue='Sales Volume')
```

```python
second_melted = avocado.melt(
    id_vars = 'year',
    value_vars=['Small Bags','Large Bags','XLarge Bags'],
    var_name ='Bags',
    value_name='Count')

with pd.option_context('display.max_rows', 10):
    display(second_melted)
```

```python
sns.relplot(data=second_melted, kind='line', x='year', y='Count', hue='Bags')
```
